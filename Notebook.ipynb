## 2. Sleigh Toy [3D]

This example loads a standard 3D Gaussian Splatting PLY file (`3d_sleigh_toy.ply`). Since it's 3D data, the splats will be static (no motion over time). This tests the renderer is correctly implemented and can handle traditional Gaussian Splatting scenes.

The cell below loads the data and converts each splat into a `Splat4D` object with zero motion.

```python
model = load_ply_3d('data/3d_sleigh_toy.ply')
from tqdm import tqdm

print('Loading gaussians ...')
gaussian3D_sleigh_objs = []


for i, (xyz, scale, rot, opacity, sh) in enumerate(tqdm(zip(model.xyz, model.scale, model.rot, model.opacity, model.sh))):
    # Convert the zeroth-order Spherical Harmonic (SH) coefficients to a [0, 1] RGB color
    # Quick conversion from sh to dc (diffuse color) only using the zeroth-order of spherical harmonics
    # to get view-independent base color of the splat (shininess and iridescence are discarded)
    dc = sh[0:3]
    motion = np.zeros(9, dtype=np.float32)  # Motion vector (9D)
    omega = np.zeros(4, dtype=np.float32)  # Angular velocity (4D)
    trbf_center = np.array([1.0], dtype=np.float32)  # Temporal radial basis function center (1D)
    trbf_scale = np.array([1.0], dtype=np.float32)  # Temporal radial basis function scale (1D)

    # --- For debugging: Print the first sample ---
    if i == 0:
        print("\n--- Data for the first Splat4D object ---")
        print(f"xyz: {xyz} (shape: {xyz.shape}, dtype: {xyz.dtype})")
        print(f"rot: {rot} (shape: {rot.shape}, dtype: {rot.dtype})")
        print(f"scale: {scale} (shape: {scale.shape}, dtype: {scale.dtype})")
        print(f"dc: {dc} (shape: {dc.shape}, dtype: {dc.dtype})")
        print(f"opacity: {opacity} (shape: {opacity.shape}, dtype: {opacity.dtype})")
        print(f"motion: {motion} (shape: {motion.shape}, dtype: {motion.dtype})")
        print(f"omega: {omega} (shape: {omega.shape}, dtype: {omega.dtype})")
        print(f"trbf_center: {trbf_center} (shape: {trbf_center.shape}, dtype: {trbf_center.dtype})")
        print(f"trbf_scale: {trbf_scale} (shape: {trbf_scale.shape}, dtype: {trbf_scale.dtype})")
        print("------------------------------------------\n")

    gaussian3D_sleigh_objs.append(Splat4D(xyz, rot, scale, dc, opacity, motion, omega, trbf_center, trbf_scale))
    
```

```python
camera = Camera(h=720, 
                w=1280, 
                position=(-0.57651054, 2.99040512, -0.03924271), 
                target=(-0.0, 0.0, 0.0))

renderer = Renderer(
    gauss_objs=gaussian3D_sleigh_objs,
    camera=camera,
    num_frames=100
)

renderer.render_frame(frame_num=1)
plt.show()
```
